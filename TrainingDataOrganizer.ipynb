```python
import os
import json
from SciServer import CasJobs
import shutil
#Consider how to detect duplicates in training set? or re-uploads of same data

folder_to_process = '/home/idies/workspace/paradim_data/pheobes_data/PARADIM/4/data_frames' # folder containing unprocessed/unsorted training data of a particular class
current_class = 'fastTop' #current class of data in folder to process
dataset_folder = '/home/idies/workspace/paradim_data/MRCNN_training_data/boron_carbide/train/fastTop' #folder in paradim to move data to
my_context = 'MyDB'

create_table_query = '''
CREATE TABLE boron_carbide_training
(
Id INT NOT NULL IDENTITY PRIMARY KEY,
json_path varchar(150),
frame_path varchar(150),
class varchar(15)
)
'''
#CasJobs.executeQuery(sql=create_table_query, context=my_context)

def getNextID():
    largest_ID_query = 'Select max(Id) from boron_carbide_training'
    id_df = CasJobs.executeQuery(sql=largest_ID_query, context=my_context)
    if id_df['Column1'][0] == None:
        return 1
    else:
        return id_df['Column1'][0] + 1

    
    
#query DB for largest ID - to use in renaming the data
cur_ID = getNextID()
#iterate over files in a folder
    #if .json
        #rename json file, image file to <class><ID>
        #read json, rewrite imagePath field to new image file name
        #move newly named json and image file to specified dataset directory in paradim volume
        #upload ID, json path, image file path, training class, maybe date time to CasJobs table
    #else 
        #continue
        
for cur_json in os.listdir(folder_to_process):
    if '.json' in cur_json:
        new_json_fn = '{0}{1}.json'.format(current_class, cur_ID)
        new_frame_fn = '{0}{1}.jpg'.format(current_class, cur_ID)
        
        new_json_path = os.path.join(dataset_folder, new_json_fn)
        new_frame_path = os.path.join(dataset_folder, new_frame_fn)
        
        annotations = json.load(open(os.path.join(folder_to_process, cur_json)))
        cur_frame = annotations['imagePath']
        annotations['imagePath'] = new_frame_fn
        
        #copy edited json to new dataset location
        with open(new_json_path, 'w') as dumpfile:
            json.dump(annotations, dumpfile)
        
        #copy frame to new dataset location
        shutil.copyfile(os.path.join(folder_to_process, cur_frame), new_frame_path)
        
        insert_query = '''insert into boron_carbide_training values ('{1}', '{2}', '{3}') '''.format(cur_ID, new_json_path, new_frame_path, current_class)
        print(insert_query)
        CasJobs.executeQuery(sql=insert_query, context=my_context)
        
        cur_ID = cur_ID + 1
    else:
        continue
```
